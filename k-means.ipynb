```python
from math import sqrt
def k_means(x_coords,y_coords,k):
    coords = list(zip(x_coords,y_coords))
    painted_coords = []
    ## Pick two random centers
    centers = zip(random.sample(coords,k),list(range(k)))
    
    ## Foreach coordinate
    for x,y in coords:
        if (x,y) not in centers:
            ## Find the closest center
            min_dx,min_dy = centers[0]
            min_d = euc_dist(x,y,min_dx,min_dy)
            
            for center_x,center_y in centers:
                d = euc_dist(x,y,center_x,center_y)
                if d<min_d:
                    min_d,min_dx,min_dy = d,center_x,center_y
    
'''
Euclidean Distance
'''
def euc_dist(x1,y1,x2,y2):
    return sqrt((x1-x2)**2+(y1-y2)**2)
        
    
```
